```python
# import dependencies
import yfinance as yf
import numpy as np
```

```python
# list of ticker names, should be able to pull this from outside somehow
tnames = ['SPY', 'TSLA']

# use yfinance function to generate list of callable tickers
tickers = [yf.Ticker(stock) for stock in tnames]
```

```python
# create list of stock price history for chosen period of time
histlist = [tick.history(period='1mo') for tick in tickers]
```

```python
len(histlist)
```

```python
# We could construct a dictionary of tickers and stock history dataframes
hist_dict = {}
indexval = np.arange(0, len(histlist), 1)

for ind in indexval:
    hist_dict[tnames[ind]] = histlist[ind]
```

```python
hist_dict['TSLA']
```

```python
# loop through dictionary and add columns using .values() suffix
for df in hist_dict.values():
    df['Amp'] = df['High'] - df['Low']
```

```python
hist_dict['TSLA']
```

Or we could keep it as list of tickers and lists of dataframes and do things like this

```python
# loop through all history dataframes in list, adding columns as needed to each one
# for example, add ticker and range between high and low point for the day
for ind in indexval:
    histlist[ind]['Amp'] = histlist[ind]['High'] - histlist[ind]['Low']
    histlist[ind]['Ticker'] = tnames[ind]
```
