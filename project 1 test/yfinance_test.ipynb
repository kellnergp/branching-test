```python
# import dependencies
import yfinance as yf
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# list of ticker names, should be able to pull this from outside somehow
tnames = ['SPY', 'TSLA']

# use yfinance function to generate list of callable tickers
tickers = [yf.Ticker(stock) for stock in tnames]
```

```python
# create list of stock price history for chosen period of time
histlist = [tick.history(period='1mo') for tick in tickers]
```

```python
len(histlist)
```

```python
# We could construct a dictionary of tickers and stock history dataframes
hist_dict = {}
indexval = np.arange(0, len(histlist), 1)

for ind in indexval:
    hist_dict[tnames[ind]] = histlist[ind]
```

```python
hist_dict['TSLA']
```

```python
# loop through dictionary and add columns using .values() suffix
for df in hist_dict.values():
    df['Amp'] = df['High'] - df['Low']
```

```python
len(hist_dict['SPY'])
```

```python
# gonna try having it loop through tickers plotting made-up mention values against close price
randomlist = []
counter = 1

# first, generate some random values to use in place of mentions
for i in range(0,23):
    n = random.randint(1, 30)
    randomlist.append(n)

# loop through tickers, creating a fake 'mentions' column then plotting them against closing value and saving the plot   
for df in hist_dict.values():
    df['Mentions'] = randomlist
    df.plot.scatter(x='Mentions', y='Close')
    plt.savefig(f"fig{str(counter)}.png")
    counter +=1
    
```
